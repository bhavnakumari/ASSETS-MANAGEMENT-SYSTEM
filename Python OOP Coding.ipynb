### 26th Jan IPA - Gold Invoice                                                  
#### (IPA) - ION PROCTURED ASSESMENT

```python
class GoldInvoice:
    def __init__(self,item_id,item_name,item_qty,item_rate,item_weight,item_pwc,item_pdis):
        self.item_id=item_id
        self.item_name=item_name
        self.item_qty=item_qty
        self.item_rate=item_rate
        self.item_weight=item_weight
        self.item_pwc=item_pwc
        self.item_pdis=item_pdis
def calc_item_price_exGST(goldInvoice_list,item_id_input):
    for i in goldInvoice_list:
        if(i.item_id == item_id_input):
            item_cost = (i.item_qty * i.item_weight * i.item_rate)
            pwc = item_cost * ( i.item_pwc / 100 )  #percentage of wastage
            pdis = item_cost * ( i.item_pdis / 100 ) #percentage of discharge
            total_cost = ( item_cost + pwc ) - pdis
        return total_cost
    return None    
def calc_item_price_inGST(goldInvoice_list,item_id_input):
    for i in goldInvoice_list:
        if(i.item_id == item_id_input):
            item_cost = (i.item_qty * i.item_weight * i.item_rate)
            pwc = item_cost * ( i.item_pwc / 100 )  #percentage of wastage
            pdis = item_cost * ( i.item_pdis / 100 ) #percentage of discharge
            gst = ( item_cost * (3 / 100 ) )
            total_cost =( ( item_cost + pwc ) - pdis ) + gst
        return total_cost  
    return None

if __name__ == '__main__':
    goldInvoice_list=[]
    goldInvoice_Count=int(input("goldInvoice_Count: "))
    for i in range(goldInvoice_Count):
        gold_item_id=int(input("gold_item_id: "))
        gold_item_name=input("gold_item_name: ")
        gold_item_qty=float(input("gold_item_qty: "))
        gold_item_rate=float(input("gold_item_rate: "))
        gold_item_weight=float(input("gold_item_weight: "))
        gold_item_pwc=float(input("gold_item_pwc: "))
        gold_item_pdis=float(input("gold_item_pdis: "))
        g=GoldInvoice(gold_item_id,gold_item_name,gold_item_qty,gold_item_rate,
                      gold_item_weight,gold_item_pwc,gold_item_pdis)
        goldInvoice_list.append(g)
    
    item_id_input=int(input("item_id_input: "))
    
    func1=calc_item_price_exGST(goldInvoice_list,item_id_input)
    if ( func1 == None ):
        print("No Item Found")
    else:
        print("{:.3f}".format(func1))
        
    func2=calc_item_price_inGST(goldInvoice_list,item_id_input)
    if ( func1 == None ):
        print("No Item Found")
    else:
        print("{:.2f}".format(func2))    
    
```

### 12the FEB OPA (PASSENGER)
#### (OPA) - ONLINE PROCTURED ASSESMENT

```python
class Passenger:
    def __init__(self,passengerName,passengerAge,distanceTravelled):
        self.passengerName=passengerName
        self.passengerAge=passengerAge
        self.distanceTravelled=distanceTravelled
def calculateTicketFare(passenger_list,fare_per_kilometer):
    tot_ticketFare=0
    for i in passenger_list:
        if (i.passengerAge<12 or i.passengerAge>=60):
            tax=0
            ticketFare=(i.distanceTravelled*fare_per_kilometer)+tax
            tot_ticketFare=tot_ticketFare+ticketFare
            
        elif (i.passengerAge>=21 and i.passengerAge<=59):
            ticket_fare_without_tax=(i.distanceTravelled*fare_per_kilometer)
            tax=(ticket_fare_without_tax*12)/100
            ticketFare=ticket_fare_without_tax+tax
            tot_ticketFare=tot_ticketFare+ticketFare
            
        else:
            ticket_fare_without_tax=(i.distanceTravelled*fare_per_kilometer)
            tax=(ticket_fare_without_tax*10)/100
            ticketFare=ticket_fare_without_tax+tax
            tot_ticketFare=tot_ticketFare+ticketFare
    return tot_ticketFare
        
def countSeniorJuniorPassengers(passenger_list):
    count=0
    for i in passenger_list:
        if (i.passengerAge<12 or i.passengerAge>=60):
            count=count+1
    return count
    
if __name__ == '__main__':
    passenger_count=int(input("passenger_count: "))
    passenger_list=[]
    for i in range(passenger_count):
        passenger_name=input("passenger_name: ")
        passenger_age=int(input("passenger_age: "))
        distance_travelled=int(input("distance_travelled: "))
        passen=Passenger(passenger_name,passenger_age,distance_travelled)
        passenger_list.append(passen)
    fare_per_kilometer=int(input("fare_per_kilometer: "))
    method1=calculateTicketFare(passenger_list,fare_per_kilometer)
    print(method1)
    method2=countSeniorJuniorPassengers(passenger_list)
    print(method2)
```

### 25TH FEB - IRC & DRC (Player & Team)
#### (IRC) - INFRASTRUCTURE READINESS CHECK
#### (DRC) - DRY RUN CHECK

```python
class Player:
    def __init__(self,playerName,playerCountry,playerAge,noOfMatches,noOfRuns,noOfWickets):
        self.playerName=playerName
        self.playerCountry=playerCountry
        self.playerAge=playerAge
        self.noOfMatches=noOfMatches
        self.noOfRuns=noOfRuns
        self.noOfWickets=noOfWickets

class Team:
    def __init__(self,list_of_players):
        self.list_of_players=list_of_players
    def getMinRuns(self):
        runs=[]
        for i in self.list_of_players:
            runs.append(i.noOfRuns)
        minimum=min(runs)
        answer1=[]
        for i in self.list_of_players:
            if (i.noOfRuns==minimum):
                answer1.append(i.playerName)
                answer1.append(i.noOfRuns)
                answer1.append(i.playerCountry)
        return answer1            
            
    def getMaxWickets(self):
        wickets=[]
        for i in self.list_of_players:
            wickets.append(i.noOfWickets)
        maximum=max(wickets)
        answer2=[]
        for i in self.list_of_players:
            if (i.noOfWickets==maximum):
                answer2.append(i.playerName)
                answer2.append(i.noOfWickets)
                answer2.append(i.playerCountry)
        return answer2
if __name__ == '__main__':
    list_of_players=[]
    player_count=int(input())
    for i in range(player_count):
        player_name=input()
        player_country=input()
        player_age=int(input())
        no_of_matches=int(input())
        no_of_runs=int(input())
        no_of_wickets=int(input())
        
        list_of_players.append(Player(player_name,player_country,player_age,no_of_matches,no_of_runs,no_of_wickets))
    o1=Player(player_name,player_country,player_age,no_of_matches,no_of_runs,no_of_wickets)
    o2=Team(list_of_players)
    ans1=o2.getMinRuns()
    for i in range(len(ans1)):
        print(ans1[i])

    ans2=o2.getMaxWickets()
    for i in range(len(ans2)):
        print(ans2[i])        
        
```
